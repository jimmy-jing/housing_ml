### Data Cleaning & Imputation

```python
for col in ames.columns:
    print("----- %s -----" %col)
    print(ames[col].value_counts())
    print(ames[col].dtype)
    print(str(sum(pd.isna(ames[col]))) + " NA VALUES")
```
