this is the AIC feature set, y has not been transformed. lets test on lasso and ridge.

```python
from sklearn.model_selection import KFold, train_test_split
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error
from math import exp
```

```python
random_state=24
k6 = KFold(n_splits=6, shuffle=True, random_state=random_state)

train = []
test = []

for train_idx, val_idx in k6.split(aic_df):
    train.append(train_idx)
    test.append(val_idx)

ridge = Ridge(alpha=2, random_state=random_state, max_iter=10000)
lasso = Lasso(alpha=36, random_state=random_state, max_iter=10000)
lasso56 = Lasso(alpha=56, random_state=random_state, max_iter=10000)

score_log_aic = []
m=1
for model in (ridge,lasso, lasso56):

    for i in range(6):
        single_log = []

        X_train = aic_df.iloc[train[i],:-1]
        X_test = aic_df.iloc[test[i],:-1]
        y_train = aic_df.iloc[train[i],-1]
        y_test = aic_df.iloc[test[i],-1]


        model.fit(X_train,y_train) 
        single_log.append(i)
        single_log.append(m)
        single_log.append(model.score(X_train,y_train))
        single_log.append(model.score(X_test, y_test))
        single_log.append(mean_squared_error(y_test, model.predict(X_test))**0.5)

        score_log_aic.append(single_log)
    m+=1
score_df_aic = pd.DataFrame(score_log_aic, columns=['Fold','Regressor','TrainScore','TestScore','MSE'])


    
        
    
    
    
```

```python
random_state=38
k6 = KFold(n_splits=6, shuffle=True, random_state=random_state)

train = []
test = []

for train_idx, val_idx in k6.split(aic_df):
    train.append(train_idx)
    test.append(val_idx)

ridge = Ridge(alpha=2, random_state=random_state, max_iter=10000)
lasso = Lasso(alpha=36, random_state=random_state, max_iter=10000)
lasso56 = Lasso(alpha=56, random_state=random_state, max_iter=10000)

score_log_aic_logy = []
m=1
for model in (ridge,lasso, lasso56):

    for i in range(6):
        single_log = []

        X_train = aic_df.iloc[train[i],:-1]
        X_test = aic_df.iloc[test[i],:-1]
        y_train = np.log1p(aic_df.iloc[train[i],-1])
        y_test = np.log1p(aic_df.iloc[test[i],-1])


        model.fit(X_train,y_train) 
        single_log.append(i)
        single_log.append(m)
        single_log.append(model.score(X_train,y_train))
        single_log.append(model.score(X_test, y_test))
        
        new_y_pred = []
        for j in model.predict(X_test):
            new_y_pred.append(exp(j))
            
        single_log.append(mean_squared_error(aic_df.iloc[test[i],-1], new_y_pred)**0.5)

        score_log_aic_logy.append(single_log)
    m+=1
score_df_aic_logy = pd.DataFrame(score_log_aic_logy, columns=['Fold','Regressor','TrainScore','TestScore','MSE'])


    
        
    
    
    
```
