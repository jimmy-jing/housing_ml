### Data Cleaning & Imputation

First identify all NA's within all columns and get an understanding of value_counts and dtype

```python
for col in ames.columns:
    print("----- %s -----" %col)
    print(ames[col].value_counts())
    print(ames[col].dtype)
    print(str(sum(pd.isna(ames[col]))) + " NA VALUES")
```

#### Lot Frontage

```python
ames.LotFrontage.fillna(0, inplace = True) #impute 0's for 259 NA's
```

#### Alley

```python
ames.Alley.fillna('NoAll', inplace = True) #input 'NoAll' for 1369 NA's
```

#### Masonry Veneer

```python
ames[pd.isnull(ames.MasVnrType)] #inspect 8 NA's from 'MasVnrType' & 'MasVnrArea' to see if they match
```

```python
ames.MasVnrType.fillna('None', inplace = True) #They do match! Impute MasVnrType to 'None'
```

```python
ames.MasVnrArea.fillna(0, inplace = True) #Impute MasVnrArea to '0' for consistency
```

#### Basement

**Note** There are 37 NA's for 'BsmtQual', 'BsmtCond', 'BsmtFinType1' which are correct & consistently labeled.


**HOWEVER** 'BsmtFinType2', 'BsmtExposure' have 38 each, so there is 1 incorrect NaN each: **HOW TO HANDLE?**

```python
ames[pd.isnull(ames.BsmtExposure)]
```

indx 949 has NaN BsmtExposure but a basement condition/qual with 0 SF. Will impute as 'No'

```python
ames.loc[ames.index==949, 'BsmtExposure'] = 'No'
```

```python
ames[pd.isnull(ames.BsmtFinType2)]
```

```python
ames.loc[ames.index == 333]
```

indx 333 has NaN BsmtFinType2, but other potentially dependent features:
    - Foundation: 'PConc'
    - BsmtExposure: 'No'
    - BsmtFinType1: 'GLQ'
Let's isolate and group other like-rows to hopefully find a pattern in BsmtFinType2

```python
Bsmt_grp = ames.groupby(['Foundation', 'BsmtExposure','BsmtFinType1', 'BsmtFinType2'])
```

```python
Bsmt_grp.count()
```

Of basements with Foundation: 'PConc', BsmtExposure: 'No', and BsmtFinType1: 'GLQ'; 149 of 153 BsmtFinType2's are 'Unf'. PLUS BsmtUnfSF = 1603. Will impute idx 333 as 'Unf'.

```python
ames.loc[ames.index == 333, 'BsmtFinType2'] = 'Rec'
```

Now let's impute the other Bsmt values that genuinely have no basement as 'NB' (No Basement)

```python
BsmtNaNs = ames[ames['BsmtFinType2'].isnull()].index
```

```python
ames.loc[BsmtNaNs, 'BsmtQual'] = 'NB'
ames.loc[BsmtNaNs, 'BsmtCond'] = 'NB'
ames.loc[BsmtNaNs, 'BsmtExposure'] = 'NB'
ames.loc[BsmtNaNs, 'BsmtFinType1'] = 'NB'
ames.loc[BsmtNaNs, 'BsmtFinType2'] = 'NB'
```

#### Electrical

**Note** There is one NA in Electrical. **HOW TO HANDLE?**

```python
ames[pd.isnull(ames.Electrical)]
```

```python
ames['Electrical'].value_counts()
```

Let's see the counts of combinations on potentially Electrical-related variables.

```python
Elec_group = ames.groupby(['Heating', 'CentralAir', 'Electrical'])
```

```python
Elec_group.count()
```

Most common Electrical type with Heating as 'GasA' & CentralAir as 'Y' is 'SBrkr' by far. We will impute this.

```python
ames.loc[ames.index == 1380,'Electrical'] = 'SBrkr'
```

#### Checkpoint: What's left?

```python
ames.isnull().sum()
```

**Checkpoint Note**: Fireplace (690) NA's, Garage (81) NA's, Pool (1453) NA's, Fence (1179) NA's, & MiscFeature (1406) NA's.

#### Fireplace

```python
ames[['FireplaceQu', 'Fireplaces']][ames.FireplaceQu.isnull()]
```

These are genuinely null. Impute as 'NF' (No Fireplace)

```python
ames.loc[ames['FireplaceQu'].isnull(), 'FireplaceQu'] = 'NF'
```

#### Garage

```python
ames[ames['GarageType'].isnull()]
```

GarageType,GarageYrBlt,GarageFinish,GarageQual,GarageCond are all (81) consistently NaN with 0 Garage Area. Impute 'NG' for No Garage.

```python
GrgNans = ames[ames['GarageType'].isnull()].index
```

```python
ames.loc[GrgNans, 'GarageType'] = 'NG'
ames.loc[GrgNans, 'GarageYrBlt'] = 'NG'
ames.loc[GrgNans, 'GarageFinish'] = 'NG'
ames.loc[GrgNans, 'GarageQual'] = 'NG'
ames.loc[GrgNans, 'GarageCond'] = 'NG'
```

**Future Note** may have trouble with 'GarageYrBlt' later in formatting dates.

#### Pool

```python
ames[['PoolQC', 'PoolArea']][ames['PoolQC'].isnull()]
```

PoolQC NaN's are truely No Pool. Impute with 'NP'.

```python
ames.loc[ames.PoolQC.isnull(), 'PoolQC'] = 'NP'
```

#### Fence

```python
ames[ames.Fence.isnull()]
```

From description file: NA in 'Fence' means No Fence. No potentially dependent variables to check against. Impute to 'NF' for No Fence.

```python
ames.loc[ames.Fence.isnull(), 'Fence'] = 'NF'
```

#### MiscFeature

If MiscFeature's are truely NA, they will have MiscVal of $0. Let's check.

```python
ames[['MiscFeature', 'MiscVal']][ames.MiscFeature.isnull()]
```

Confirmed MiscVal of 0 for all NA MiscFeatures. Impute as NMF (No Misc Features).

```python
ames.loc[ames.MiscFeature.isnull(), 'MiscFeature'] = 'NMF'
```
